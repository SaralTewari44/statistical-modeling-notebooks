```python
import pandas as pd
from sklearn.metrics import confusion_matrix, precision_score, accuracy_score, recall_score, f1_score
import matplotlib.pyplot as plt
import seaborn as sns

```

```python
data = pd.read_excel("HW4 - Model Evaluation - Data.xlsx")
data.head()
```

```python
y_pred = data['Predicted']
y_true = data['Actual']

#the confusion matrix
cm = confusion_matrix(y_true, y_pred, labels=["Dog", "Not Dog"])

# plotting it out
plt.figure(figsize=(6, 4))
sns.heatmap(cm, annot=True, fmt='g', cmap="Blues",
            xticklabels=["Dog", "Not Dog"],
            yticklabels=["Dog", "Not Dog"])

plt.xlabel("Predicted")
plt.ylabel("Actual")
plt.title("Confusion Matrix (Excel Data)")
plt.show()
```

```python
precision = precision_score(y_true, y_pred, pos_label="Dog")
accuracy = accuracy_score(y_true, y_pred)
recall = recall_score(y_true, y_pred, pos_label="Dog")
f1 = f1_score(y_true, y_pred, pos_label="Dog")

print("Total Sample Size:", len(y_true))
print("Precision:", round(precision, 2))
print("Accuracy:", round(accuracy, 2))
print("Recall:", round(recall, 2))
print("F-measure:", round(f1, 2))
```

```python
# list of car brands
y_true = ["honda", "kia", "honda", "toyota", "kia", "chevrolet"]

# model prediction
y_pred = ["honda", "kia", "honda", "toyota", "toyota", "honda"]
```

```python
labels = ["honda", "kia", "toyota", "chevrolet"]
cm = confusion_matrix(y_true, y_pred, labels=labels)

plt.figure(figsize=(6, 4))
sns.heatmap(cm, annot=True, fmt='g', cmap="Greens",
            xticklabels=labels, yticklabels=labels)

plt.xlabel("Predicted")
plt.ylabel("Actual")
plt.title("Confusion Matrix (Car Brands)")
plt.show()
```

```python
precision = precision_score(y_true, y_pred, average='macro', zero_division=0)
accuracy = accuracy_score(y_true, y_pred)
recall = recall_score(y_true, y_pred, average='macro', zero_division=0)
f1 = f1_score(y_true, y_pred, average='macro', zero_division=0)

print("Total Sample Size:", len(y_true))
print("Precision:", round(precision, 2))
print("Accuracy:", round(accuracy, 2))
print("Recall:", round(recall, 2))
print("F-measure:", round(f1, 2))
```
